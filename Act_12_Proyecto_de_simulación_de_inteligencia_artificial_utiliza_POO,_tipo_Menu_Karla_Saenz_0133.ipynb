```python
# --------------------------------------------------------
# PROGRAMA: SIMULADOR INTELIGENTE CON POO (CLASE: Influencer)
# --------------------------------------------------------


# Definimos la clase en singular como se solicitó
class Influencer:
    # El método constructor inicializa los atributos de cada influencer
    def __init__(self, nombre, nicho, plataforma, engagement, colaboraciones, seguidores, frecuencia):

        self.nombre = nombre
        self.nicho_de_mercado = nicho
        self.plataforma_principal = plataforma
        self.tasa_engagement = engagement
        self.colaboraciones_de_marca = colaboraciones
        self.numero_seguidores = seguidores
        self.frecuencia_posteo = frecuencia


# Clase que gestionará el comportamiento, la "IA" y el menú (sin usar POO compleja)
class SistemaInfluencer:

    def __init__(self):
        # Lista vacía que guardará los OBJETOS de tipo Influencer
        self.lista_influencer = []

    def agregar_medico(self):
        # Nota para los alumnos: Se mantiene el nombre de la función asignada por el ejercicio
        print("\n--- 📝 REGISTRAR DATOS DEL INFLUENCER ---")

        # Entrada de datos interactiva
        nombre = input("👉 Nombre del influencer: ")
        nicho = input("👉 Nicho de mercado (ej. Gamer, Moda, Tech): ")
        plataforma = input("👉 Plataforma principal (ej. TikTok, Instagram): ")
        engagement = input("👉 Tasa de engagement (ej. 4.5%): ")
        colaboraciones = input("👉 Colaboraciones de marca (Sí/No): ")
        seguidores = input("👉 Número de seguidores (ej. 100K): ")
        frecuencia = input("👉 Frecuencia de posteo (ej. Diario, Semanal): ")

        # CREACIÓN DEL OBJETO: Pasamos los datos al constructor de la clase Influencer
        nuevo_influencer = Influencer(
            nombre, nicho, plataforma, engagement, colaboraciones, seguidores, frecuencia
        )

        # Guardamos el objeto en nuestra lista
        self.lista_influencer.append(nuevo_influencer)
        print(f"\n✅ ¡Influencer '{nombre}' agregado exitosamente como objeto!")

    def listar_medicos(self):
        print("\n--- 📊 TABLA DE INFLUENCER (VISTA DE OBJETOS) ---")

        if len(self.lista_influencer) == 0:
            print("No hay ningún influencer registrado en el sistema.")
        else:
            # Encabezado de la tabla con distribución alineada
            print(
                f"{'Nombre':<15} | {'Nicho':<12} | {'Plataforma':<12} | {'Engagement':<10} | {'Colaboración':<12} | {'Seguidores':<10} | {'Frecuencia':<12}"
            )
            print("-" * 95)

            # Recorremos la lista y accedemos a los ATRIBUTOS de cada objeto usando el punto (.)
            for inf in self.lista_influencer:
                print(
                    f"{inf.nombre:<15} | {inf.nicho_de_mercado:<12} | {inf.plataforma_principal:<12} | {inf.tasa_engagement:<10} | {inf.colaboraciones_de_marca:<12} | {inf.numero_seguidores:<10} | {inf.frecuencia_posteo:<12}"
                )

    def actualizar_medico(self):
        print("\n--- 🔄 ACTUALIZAR DATOS DE INFLUENCER ---")
        if len(self.lista_influencer) == 0:
            print("No hay registros para actualizar.")
            return

        nombre_buscar = input(
            "Introduce el nombre del influencer que deseas modificar: "
        )
        encontrado = False

        for inf in self.lista_influencer:
            if inf.nombre.lower() == nombre_buscar.lower():
                print(f"\n¡Influencer '{inf.nombre}' encontrado! Ingresa los nuevos datos:")
                inf.nicho_de_mercado = input("👉 Nuevo Nicho de mercado: ")
                inf.plataforma_principal = input("👉 Nueva Plataforma principal: ")
                inf.tasa_engagement = input("👉 Nueva Tasa de engagement: ")
                inf.colaboraciones_de_marca = input("👉 Nuevas Colaboraciones (Sí/No): ")
                inf.numero_seguidores = input("👉 Nuevo Número de seguidores: ")
                inf.frecuencia_posteo = input("👉 Nueva Frecuencia de posteo: ")
                print("\n✅ ¡Datos actualizados correctamente!")
                encontrado = True
                break

        if not encontrado:
            print("❌ No se encontró ningún influencer con ese nombre.")

    def borrar_medico(self):
        print("\n--- 🗑 BORRAR INFLUENCER ---")
        if len(self.lista_influencer) == 0:
            print("No hay registros para borrar.")
            return

        nombre_buscar = input("Introduce el nombre del influencer a eliminar: ")
        encontrado = False

        for inf in self.lista_influencer:
            if inf.nombre.lower() == nombre_buscar.lower():
                self.lista_influencer.remove(inf)
                print(f"\n🗑 ¡El influencer '{inf.nombre}' ha sido eliminado del sistema!")
                encontrado = True
                break

        if not encontrado:
            print("❌ No se encontró ningún influencer con ese nombre.")

    def simular_ia_analisis(self):
        """Esta función simula la IA analizando los objetos creados"""
        print("\n--- 🤖 ANÁLISIS INTELIGENTE DE DATOS (IA) ---")
        if len(self.lista_influencer) == 0:
            print("La IA no tiene objetos para analizar de manera predictiva.")
            return

        print("Optimizador de campañas🤖:")
        print("La IA buscará perfiles con potencial viral (Frecuencia Diaria o Alta).")
        print("\n[IA escaneando los atributos de la clase Influencer...]")

        encontrado = False
        for inf in self.lista_influencer:
            # La IA evalúa las propiedades del objeto directamente
            if inf.frecuencia_posteo.lower() == "diario" or "alta" in inf.frecuencia_posteo.lower():
                print(f"🔥 ALERTA IA: ¡{inf.nombre} es un perfil altamente recomendado para pautas comerciales!")
                encontrado = True

        if not encontrado:
            print("🤖 IA: Tras analizar los objetos, ninguno cumple con el criterio de alta frecuencia actual.")


# --- CICLO PRINCIPAL (MENU INTERACTIVO) ---
# Instanciamos el sistema que controla todo
sistema = SistemaInfluencer()

while True:
    print("\n==========================================")
    print("  💻 SISTEMA POO - CLASE: Influencer (IA) ")
    print("==========================================")
    print("1. ➕ Agregar Influencer         (agregar_medico)")
    print("2. 📊 Listar todos en Tabla      (listar_medicos)")
    print("3. 🔄 Actualizar datos           (actualizar_medico)")
    print("4. 🗑 Borrar registro            (borrar_medico)")
    print("5. 🤖 Ejecutar Diagnóstico IA")
    print("6. ❌ Salir del programa")

    opcion = input("\nSelecciona una opción del menú (1-6): ")

    # Estructura de control IF-ELIF-ELSE para el flujo del programa
    if opcion == "1":
        sistema.agregar_medico()
    elif opcion == "2":
        sistema.listar_medicos()
    elif opcion == "3":
        sistema.actualizar_medico()
    elif opcion == "4":
        sistema.borrar_medico()
    elif opcion == "5":
        sistema.simular_ia_analisis()
    elif opcion == "6":
        print("\n👋 Saliendo del simulador de clase. ¡Buen código!")
        break
    else:
        print("\n⚠ Opción no válida. Por favor, selecciona un número del 1 al 6.")
```
